## Scatterplot Matrisi

```python
iris= sns.load_dataset("iris")
df=iris.copy()
df.head()
```

```python
sns.pairplot(df)
## bize iki bilgi sunar ilki eksenlerdeki grafikler ile değişkenlerin dağılımları gösterilir
## ikincisi ise değişkenlerin birbirleri ile olan ilişkileridir
## eğer grafik toz bulutu şeklinde ise veya yapısal bir formu yoksa bu iki değişken arasında ilişki yoktur demektir
## gözlem grupları eğer birbirlerinden farklı köşelerde kümeleniyorlarsa çaprazlama durumu ile gözlemleyebileceğimiz bazı alt gruplar vardır
```

```python
sns.pairplot(df,hue="species")
```

```python
sns.pairplot(df,hue="species",markers=["o","s","D"])
```

```python
sns.pairplot(df,kind="reg",hue="species",markers=["d","o","s"])
```
