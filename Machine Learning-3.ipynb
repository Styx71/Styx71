```python
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import Ridge,RidgeCV,Lasso,LassoCV
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("reading_data/Hitters.csv")
df = df.dropna()
df = pd.get_dummies(df,["League","Division","NewLeague"],["League_dum","Division_dum","NewLeague_dum"])
```

```python
df
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(["Salary"],axis=1),df["Salary"],test_size=0.2)
model = Ridge(0.1).fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

```python
alfalar = np.linspace(10,-10,1000)
```

```python
cross_val= RidgeCV(alphas=alfalar,scoring = "neg_mean_squared_error", cv = 10,normalize =True)
```

```python
cross_val.fit(X_train,y_train)
```

```python
cross_val.alpha_
```

```python
#Now we know what is alpha supposed to be
model = Ridge(cross_val.alpha_).fit(X_train,y_train)
y_pred = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

***ELASTICNET***

```python
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import Ridge,RidgeCV,Lasso,LassoCV,ElasticNet, ElasticNetCV
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("reading_data/Hitters.csv")
df = df.dropna()
df = pd.get_dummies(df,["League","Division","NewLeague"],["League_dum","Division_dum","NewLeague_dum"])
x = df.drop("Salary",axis=1)
y = df["Salary"]
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size = 0.20)
```

```python
model = ElasticNet()
model.fit(X_train,y_train)
```

```python
cross_val = ElasticNetCV(alphas=np.linspace(-10,10,100),cv=10,normalize=True)
```

```python
cross_val.fit(X_train,y_train)
```

```python
cross_val.alpha_
```

```python
model = ElasticNet(alpha=cross_val.alpha_)
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```
